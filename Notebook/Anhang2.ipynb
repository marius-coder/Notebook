```python
print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~ ")
print("In the current dataframe df_2015 there are several rows with NAN-values ")
print("This function aims to fix these gaps by calculating a linear slope and extraploating the values (Geradengleichung) ")
print("Input:  df_2015, 'row of Dataframe','colum of Dataframe' ")
print("Output:'Fixed Dataframe'")
print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~ ")
def FixNANValues(column,dfSource):
#if datapoint is empty we replace it with the average of the datapoint before and after 
    row = 0
    rowBeg = row 
    
    while row < len(dfSource.index):
        if dfSource.iloc[row,column] != dfSource.iloc[row,column]:
            rowBeg = row
            #Check how many empty datapoints there are
            while(dfSource.iloc[row,column] != dfSource.iloc[row,column]):
                row += 1
            #If Column is completely empty we return prematurely
            if len(dfSource.index) == row:
                return 
            slope = (dfSource.iloc[row,column]-dfSource.iloc[rowBeg-1,column])/((row-rowBeg)+1)
            while(rowBeg != row):
                #Insert interpolated Data into Dataframe
                dfSource.iloc[rowBeg,column] = dfSource.iloc[rowBeg-1,column] + slope
                rowBeg += 1
        row += 1
    return   

    
```

```python
print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~ ")
print("There is currently no clean way to add the weeknumber since index.week is deprecated")
print("This function creates them with a loop and the funciton isocalendar")
print("Name: CreateWeekNum ")
print("Input: df_Source ")
print("Output: list of weeknumbers")
print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~ ")
def CreateWeekNum(df_Source):
    li_weeknum = []
    for rows in range(0,len(df_Source),1):
        li_weeknum.append(df_Source.index[rows].isocalendar()[1])
    return li_weeknum    
```

```python
def FixScale_Präs(df_Source,month):
    print("FixScale: START")
    df_Old = pd.DataFrame()
    df_Old = df_Source.copy()
 
    

   # print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
    #print("MONAT: ", month)
    #print("Before: ",sum(df_Source[str(month)]))

    var_lin_Factor = 1
    var_Factor = 0.00001

    while abs(sum(df_Source[str(month)])-100) > 0.001 :
     
        if abs(sum(df_Source[str(month)])-100) > 1:
            var_Factor = 0.01
            #print("var_Factor ist: ", var_Factor)
        elif abs(sum(df_Source[str(month)])-100) < 1:
            var_Factor = 0.00001
            #print("var_Factor ist: ", var_Factor)
    
        if sum(df_Source[str(month)]) > 100: 
            #print("var_Factor ist NEGATIV")
            var_Factor *= -1
               
        var_lin_Factor += var_Factor  
        df_Source[str(month)] = df_Old[str(month)] * var_lin_Factor
     
    #print("After: ",sum(df_Source[str(month)]))
        
    
    print("FixScale: DONE")
    return df_Source
def Modulate_Präs(df_Source,month):
    print("Modulate: START")
        
    li_Scale_exp = []  
    
    df_Scale_exp = pd.DataFrame()
    df_Scale_exp = df_Source.copy()
    
    
    
    
    var_days_in_month = pd.Period("2018-" + str(month) +"-17").days_in_month
    li_Scale_exp = []
    
    
    for hour in range(0,24,1):
        li_Scale_exp.append(li_Scale[hour])
    #print("MONTH: ", month)
    #modulate signal
    df_Scale_exp[str(month)] = df_Source[str(month)] * li_Scale_exp
    
    #for day in range(1,var_days_in_month +1,1):
        #print(df_Scale_exp[str(month)].loc[day * 24: day * 24 + 24].sum())
     
    
    
    
    
    df_Modulated = pd.DataFrame()
    
    
    df_Modulated = FixScale_Präs(df_Scale_exp,month)
    print("Modulate: DONE")
    return df_Modulated
```

```python
li_ET = ["Biomasse","Hydro","Solar","Wind","Gas","PSP",]
li_CO2 = []
li_CO2_Biomass = []
li_CO2_Hydro = []
li_CO2_Solar = []
li_CO2_Wind = []
li_CO2_Gas = []
li_CO2_PSP = []
li_CO2_H2 = []
li_CO2_Geot = []
li_CO2_Import = []


li_CO2_Biomass_percent = []
li_CO2_Hydro_percent = []
li_CO2_Solar_percent = []
li_CO2_Wind_percent = []
li_CO2_Gas_percent = []
li_CO2_PSP_percent = []
li_CO2_H2_percent = []
li_CO2_Geot_percent = []
li_CO2_Import_percent = []





for hour in range(0,8760,1):  
    
    sum_ET = df_2030["Biomasse"].iloc[hour] + df_2030["Hydro"].iloc[hour] + df_2030["Solar"].iloc[hour] + \
     df_2030["Wind"].iloc[hour] + df_2030["Gas"].iloc[hour] + \
    df_2030["PSP"].iloc[hour] + df_2030["H2"].iloc[hour] + df_2030["Geot"].iloc[hour]
    
    var_CO2 = (df_2030["Biomasse"].iloc[hour] * df_Scaling_Factors["CO2 Factors [g/kWh]"]["Biomasse"] + \
               df_2030["Hydro"].iloc[hour] * df_Scaling_Factors["CO2 Factors [g/kWh]"]["Hydro"] + \
               df_2030["Solar"].iloc[hour] * df_Scaling_Factors["CO2 Factors [g/kWh]"]["Solar"] + \
               df_2030["Wind"].iloc[hour] * df_Scaling_Factors["CO2 Factors [g/kWh]"]["Wind"] + \
               df_2030["PSP"].iloc[hour] * df_Scaling_Factors["CO2 Factors [g/kWh]"]["PSP"] + \
               df_2030["H2"].iloc[hour] * df_Scaling_Factors["CO2 Factors [g/kWh]"]["H2"] + \
               df_2030["Geot"].iloc[hour] * df_Scaling_Factors["CO2 Factors [g/kWh]"]["Geot"] + \
               df_2030["Gas"].iloc[hour] * df_Scaling_Factors["CO2 Factors [g/kWh]"]["Gas"]) / sum_ET
    li_CO2.append(var_CO2)
    
    var_CO2_Biomass = df_2030["Biomasse"].iloc[hour] * df_Scaling_Factors["CO2 Factors [g/kWh]"]["Biomasse"] / sum_ET
    li_CO2_Biomass.append(var_CO2_Biomass)
    var_CO2_Biomass_percent = var_CO2_Biomass / var_CO2 * 100
    li_CO2_Biomass_percent.append(var_CO2_Biomass_percent)
    
    var_CO2_Hydro = df_2030["Hydro"].iloc[hour] * df_Scaling_Factors["CO2 Factors [g/kWh]"]["Hydro"] / sum_ET
    li_CO2_Hydro.append(var_CO2_Hydro)
    var_CO2_Hydro_percent = var_CO2_Hydro / var_CO2 * 100
    li_CO2_Hydro_percent.append(var_CO2_Hydro_percent)
    
    var_CO2_Solar = df_2030["Solar"].iloc[hour] * df_Scaling_Factors["CO2 Factors [g/kWh]"]["Solar"] / sum_ET
    li_CO2_Solar.append(var_CO2_Solar)
    var_CO2_Solar_percent = var_CO2_Solar / var_CO2 * 100
    li_CO2_Solar_percent.append(var_CO2_Solar_percent)
    
    var_CO2_Wind = df_2030["Wind"].iloc[hour] * df_Scaling_Factors["CO2 Factors [g/kWh]"]["Wind"] / sum_ET
    li_CO2_Wind.append(var_CO2_Wind)
    var_CO2_Wind_percent = var_CO2_Wind / var_CO2 * 100
    li_CO2_Wind_percent.append(var_CO2_Wind_percent)
    
    var_CO2_PSP = df_2030["PSP"].iloc[hour] * df_Scaling_Factors["CO2 Factors [g/kWh]"]["PSP"] / sum_ET
    li_CO2_PSP.append(var_CO2_PSP)
    var_CO2_PSP_percent = var_CO2_PSP / var_CO2 * 100
    li_CO2_PSP_percent.append(var_CO2_PSP_percent)
    
    var_CO2_H2 = df_2030["H2"].iloc[hour] * df_Scaling_Factors["CO2 Factors [g/kWh]"]["H2"] / sum_ET
    li_CO2_H2.append(var_CO2_H2)
    var_CO2_H2_percent = var_CO2_H2 / var_CO2 * 100
    li_CO2_H2_percent.append(var_CO2_H2_percent)
    
    var_CO2_Geot = df_2030["Geot"].iloc[hour] * df_Scaling_Factors["CO2 Factors [g/kWh]"]["Geot"] / sum_ET
    li_CO2_Geot.append(var_CO2_Geot)
    var_CO2_Geot_percent = var_CO2_Geot / var_CO2 * 100
    li_CO2_Geot_percent.append(var_CO2_Geot_percent)
    
    var_CO2_Gas = df_2030["Gas"].iloc[hour] * df_Scaling_Factors["CO2 Factors [g/kWh]"]["Gas"] / sum_ET
    li_CO2_Gas.append(var_CO2_Gas)
    var_CO2_Gas_percent = var_CO2_Gas / var_CO2 * 100
    li_CO2_Gas_percent.append(var_CO2_Gas_percent)

    var_CO2_Import = df_2030["Import"].iloc[hour].clip(0,1e8) * df_2030["CO2_import"][hour] / \
    (sum_ET + df_2030["Import"].iloc[hour].clip(0,9e6))
    li_CO2_Import.append(var_CO2_Import)
    var_CO2_Import_percent = var_CO2_Import / (var_CO2 + df_2030["Import"].iloc[hour].clip(0,99999) * df_2030["CO2_import"][hour]) \
    * 100
    li_CO2_Import_percent.append(var_CO2_Import_percent)
               
df_2030["CO2_Produced"] = li_CO2
df_2030["CO2_abs_Biomass"] = li_CO2_Biomass
df_2030["CO2_abs_Hydro"] = li_CO2_Hydro
df_2030["CO2_abs_Solar"] = li_CO2_Solar
df_2030["CO2_abs_Wind"] = li_CO2_Wind
df_2030["CO2_abs_PSP"] = li_CO2_PSP
df_2030["CO2_abs_H2"] = li_CO2_H2
df_2030["CO2_abs_Gas"] = li_CO2_Gas
df_2030["CO2_abs_Geot"] = li_CO2_Geot
df_2030["CO2_abs_Import"] = li_CO2_Import
df_2030["CO2_Intensity"] = df_2030["CO2_Produced"] + df_2030["CO2_abs_Import"]

df_2030["CO2_percent_Biomass"] = li_CO2_Biomass_percent
df_2030["CO2_percent_Hydro"]   = li_CO2_Hydro_percent
df_2030["CO2_percent_Solar"]   = li_CO2_Solar_percent
df_2030["CO2_percent_Wind"]    = li_CO2_Wind_percent
df_2030["CO2_percent_PSP"]     = li_CO2_PSP_percent
df_2030["CO2_percent_H2"]     = li_CO2_H2_percent
df_2030["CO2_percent_Gas"]     = li_CO2_Gas_percent
df_2030["CO2_percent_Geot"] = li_CO2_Geot_percent
df_2030["CO2_percent_Import"]  = li_CO2_Import_percent
```

```python
li_ET = ["Biomasse","Hydro","Solar","Wind","Gas","PSP",]
li_CO2 = []
li_CO2_Biomass = []
li_CO2_Hydro = []
li_CO2_Solar = []
li_CO2_Wind = []
li_CO2_Gas = []
li_CO2_PSP = []
li_CO2_H2 = []
li_CO2_Geot = []
li_CO2_Import = []


li_CO2_Biomass_percent = []
li_CO2_Hydro_percent = []
li_CO2_Solar_percent = []
li_CO2_Wind_percent = []
li_CO2_Gas_percent = []
li_CO2_PSP_percent = []
li_CO2_H2_percent = []
li_CO2_Geot_percent = []
li_CO2_Import_percent = []





for hour in range(0,8760,1):  
    
    sum_ET = df_2050["Biomasse"].iloc[hour] + df_2050["Hydro"].iloc[hour] + df_2050["Solar"].iloc[hour] + \
     df_2050["Wind"].iloc[hour] + df_2050["Gas"].iloc[hour] + \
    df_2050["PSP"].iloc[hour] + df_2050["H2"].iloc[hour] + df_2050["Geot"].iloc[hour]
    
    var_CO2 = (df_2050["Biomasse"].iloc[hour] * df_Scaling_Factors["CO2 Factors [g/kWh]"]["Biomasse"] + \
               df_2050["Hydro"].iloc[hour] * df_Scaling_Factors["CO2 Factors [g/kWh]"]["Hydro"] + \
               df_2050["Solar"].iloc[hour] * df_Scaling_Factors["CO2 Factors [g/kWh]"]["Solar"] + \
               df_2050["Wind"].iloc[hour] * df_Scaling_Factors["CO2 Factors [g/kWh]"]["Wind"] + \
               df_2050["PSP"].iloc[hour] * df_Scaling_Factors["CO2 Factors [g/kWh]"]["PSP"] + \
               df_2050["H2"].iloc[hour] * df_Scaling_Factors["CO2 Factors [g/kWh]"]["H2"] + \
               df_2050["Geot"].iloc[hour] * df_Scaling_Factors["CO2 Factors [g/kWh]"]["Geot"] + \
               df_2050["Gas"].iloc[hour] * df_Scaling_Factors["CO2 Factors [g/kWh]"]["Gas"]) / sum_ET
    li_CO2.append(var_CO2)
    
    var_CO2_Biomass = df_2050["Biomasse"].iloc[hour] * df_Scaling_Factors["CO2 Factors [g/kWh]"]["Biomasse"] / sum_ET
    li_CO2_Biomass.append(var_CO2_Biomass)
    var_CO2_Biomass_percent = var_CO2_Biomass / var_CO2 * 100
    li_CO2_Biomass_percent.append(var_CO2_Biomass_percent)
    
    var_CO2_Hydro = df_2050["Hydro"].iloc[hour] * df_Scaling_Factors["CO2 Factors [g/kWh]"]["Hydro"] / sum_ET
    li_CO2_Hydro.append(var_CO2_Hydro)
    var_CO2_Hydro_percent = var_CO2_Hydro / var_CO2 * 100
    li_CO2_Hydro_percent.append(var_CO2_Hydro_percent)
    
    var_CO2_Solar = df_2050["Solar"].iloc[hour] * df_Scaling_Factors["CO2 Factors [g/kWh]"]["Solar"] / sum_ET
    li_CO2_Solar.append(var_CO2_Solar)
    var_CO2_Solar_percent = var_CO2_Solar / var_CO2 * 100
    li_CO2_Solar_percent.append(var_CO2_Solar_percent)
    
    var_CO2_Wind = df_2050["Wind"].iloc[hour] * df_Scaling_Factors["CO2 Factors [g/kWh]"]["Wind"] / sum_ET
    li_CO2_Wind.append(var_CO2_Wind)
    var_CO2_Wind_percent = var_CO2_Wind / var_CO2 * 100
    li_CO2_Wind_percent.append(var_CO2_Wind_percent)
    
    var_CO2_PSP = df_2050["PSP"].iloc[hour] * df_Scaling_Factors["CO2 Factors [g/kWh]"]["PSP"] / sum_ET
    li_CO2_PSP.append(var_CO2_PSP)
    var_CO2_PSP_percent = var_CO2_PSP / var_CO2 * 100
    li_CO2_PSP_percent.append(var_CO2_PSP_percent)
    
    var_CO2_H2 = df_2050["H2"].iloc[hour] * df_Scaling_Factors["CO2 Factors [g/kWh]"]["H2"] / sum_ET
    li_CO2_H2.append(var_CO2_H2)
    var_CO2_H2_percent = var_CO2_H2 / var_CO2 * 100
    li_CO2_H2_percent.append(var_CO2_H2_percent)
    
    var_CO2_Geot = df_2050["Geot"].iloc[hour] * df_Scaling_Factors["CO2 Factors [g/kWh]"]["Geot"] / sum_ET
    li_CO2_Geot.append(var_CO2_Geot)
    var_CO2_Geot_percent = var_CO2_Geot / var_CO2 * 100
    li_CO2_Geot_percent.append(var_CO2_Geot_percent)
    
    var_CO2_Gas = df_2050["Gas"].iloc[hour] * df_Scaling_Factors["CO2 Factors [g/kWh]"]["Gas"] / sum_ET
    li_CO2_Gas.append(var_CO2_Gas)
    var_CO2_Gas_percent = var_CO2_Gas / var_CO2 * 100
    li_CO2_Gas_percent.append(var_CO2_Gas_percent)

    var_CO2_Import = df_2050["Import"].iloc[hour].clip(0,1e8) * df_2050["CO2_import"][hour] / \
    (sum_ET + df_2050["Import"].iloc[hour].clip(0,9e6))
    li_CO2_Import.append(var_CO2_Import)
    var_CO2_Import_percent = var_CO2_Import / (var_CO2 + df_2050["Import"].iloc[hour].clip(0,99999) * df_2050["CO2_import"][hour]) \
    * 100
    li_CO2_Import_percent.append(var_CO2_Import_percent)
               
df_2050["CO2_Produced"] = li_CO2
df_2050["CO2_abs_Biomass"] = li_CO2_Biomass
df_2050["CO2_abs_Hydro"] = li_CO2_Hydro
df_2050["CO2_abs_Solar"] = li_CO2_Solar
df_2050["CO2_abs_Wind"] = li_CO2_Wind
df_2050["CO2_abs_PSP"] = li_CO2_PSP
df_2050["CO2_abs_H2"] = li_CO2_H2
df_2050["CO2_abs_Gas"] = li_CO2_Gas
df_2050["CO2_abs_Geot"] = li_CO2_Geot
df_2050["CO2_abs_Import"] = li_CO2_Import
df_2050["CO2_Intensity"] = df_2050["CO2_Produced"] + df_2050["CO2_abs_Import"]

df_2050["CO2_percent_Biomass"] = li_CO2_Biomass_percent
df_2050["CO2_percent_Hydro"]   = li_CO2_Hydro_percent
df_2050["CO2_percent_Solar"]   = li_CO2_Solar_percent
df_2050["CO2_percent_Wind"]    = li_CO2_Wind_percent
df_2050["CO2_percent_PSP"]     = li_CO2_PSP_percent
df_2050["CO2_percent_H2"]     = li_CO2_H2_percent
df_2050["CO2_percent_Gas"]     = li_CO2_Gas_percent
df_2050["CO2_percent_Geot"] = li_CO2_Geot_percent
df_2050["CO2_percent_Import"]  = li_CO2_Import_percent
```
